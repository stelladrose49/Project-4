```python
# Let's first load the uploaded dataset to understand its structure and contents
import pandas as pd

# Load the dataset
data_path = 'cleaned.csv'
data = pd.read_csv(data_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Convert price to numeric, handling '$' and ',' characters
data['Price'] = data['Price'].replace('[\$,]', '', regex=True).astype(float)

# 1. Brand vs Average Price
plt.figure(figsize=(10, 6))
sns.barplot(x='Price', y='Brand Name', data=data, estimator=np.mean, ci=None)
plt.title('Average Price by Brand')
plt.xlabel('Average Price ($)')
plt.ylabel('Brand')
plt.show()
```

```python
# 1. Brand vs RAM Distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='Brand Name', y='RAM (GB)', data=data)
plt.title('RAM Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('RAM (GB)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Distribution of Camera Max MP
plt.figure(figsize=(10, 6))
sns.histplot(data['Camera Max MP'], bins=20, kde=True)
plt.title('Distribution of Camera Max Megapixels')
plt.xlabel('Camera Max Megapixels')
plt.ylabel('Count')
plt.show()
```

```python
# 5. Distribution of Phones by Price Range
# Create price range categories
price_bins = [0, 500, 1000, 1500, 2000, 2500, 3000]
price_labels = ['0-500', '501-1000', '1001-1500', '1501-2000', '2001-2500', '2501-3000']
data['Price Range'] = pd.cut(data['Price'], bins=price_bins, labels=price_labels)

plt.figure(figsize=(10, 6))
sns.countplot(x='Price Range', data=data)
plt.title('Distribution of Phones by Price Range')
plt.xlabel('Price Range ($)')
plt.ylabel('Number of Phones')
plt.xticks(rotation=45)
plt.show()
```
